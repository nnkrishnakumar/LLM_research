```python
# Load tf_dataset of cat and horse images
# /*/* these two "/*" is for two differnt purpose first to go inside a folder and second is for file inside the function
# example and explaination of given /*/*: 
# "/*"      : >>>>> /cat images - Google Search
# "/*/*"    : >>>>> /cat images - Google Search/2_large (1).jpg

df_cat_horse=tf.data.Dataset.list_files('/home/krishnakumar/Documents/reseach/LLM_reserach/Datasets/Images_horse_cat/*/*')   
for i in df_cat_horse:
    print(i.numpy())
```

```python
# this function is use to slice the label from the folder if the image is of "cat" or "horse"
def get_label(file_path):
    import os
    # This part of the code splits the file_path using the directory separator "(os.path.sep)"", which is typically a forward slash ("/") on Unix-like systems or a backslash ("") on Windows.
    return tf.strings.split(tf.strings.split(file_path,os.path.sep)[-2])[0]     
```

```python
get_label("/home/krishnakumar/Documents/reseach/LLM_reserach/Datasets/Images_horse_cat/horse images - Google Search/1.2fb223e92647f3ce1bb19fe4212f4c04.jpg")
```

```python
# map the function get_label to get the label of the image like horse and cat using map function merging a function ::get_label() with dataset_path ::df_cat_horse
for i in df_cat_horse.map(get_label):
    print(i.numpy())
```

```python
# this function is use to read image in 3D numpy array and get the label using get_label() integrated inside the preprocess_image()
def preprocess_image(file_path):
    label=get_label(file_path)
    img=tf.io.read_file(file_path)
    img=tf.image.decode_jpeg(img)
    img=tf.image.resize(img,[128,128])
    return img,label
```

```python
for img,label in df_cat_horse.map(preprocess_image).take(3):
    print("label :",label)                     # this is the label of image
    print("IMAGE:",img[0][0])                  # because 3D array of a image contain a lot of data so just to check getting only one pixel which is in BGR or RGB channel
```

```python
for img,label in df_cat_horse.map(preprocess_image).take(3):
    print("label :",label)                     # this is the label of image
    print("IMAGE:",img)                  
```

```python
# scaling the image in between 0 to 1 deviding all the number by 255 which is uint 8 datatype.
def scale_img(image,label):
    return image/255,label
```

```python
# using map function integrate preprocess_image() and scale_img() 
for img,label in df_cat_horse.map(preprocess_image).map(scale_img).take(2):
    print("label :",label)
    print("IMAGE:",img)
```
